```python
# Import Libraries
import requests
import pandas as pd
import matplotlib.pyplot as plt

# Optional: Configure matplotlib for inline display in Jupyter
%matplotlib inline
```

## Understand the Open-Meteo API

- **API Endpoint**: `https://archive-api.open-meteo.com/v1/archive`
- **Parameters**:
  - `latitude`: Latitude of the location.
  - `longitude`: Longitude of the location.
  - `start_date`: Start date of the data (YYYY-MM-DD).
  - `end_date`: End date of the data (YYYY-MM-DD).
  - `hourly`: Comma-separated list of variables (e.g., temperature, precipitation).
  - `timezone`: Timezone of the data.

**Documentation**: [Open-Meteo API Docs](https://open-meteo.com/en/docs)

```python
# Step 1: Set Up Parameters

# Location coordinates for New York City
latitude = 40.7128
longitude = -74.0060

# Date range
start_date = '2023-01-01'
end_date = '2023-01-07'  # One week of data

# Variables to fetch
hourly_variables = ['temperature_2m', 'relativehumidity_2m', 'precipitation']
```

```python
# Step 2: Make the API Request

base_url = 'https://archive-api.open-meteo.com/v1/archive'

params = {
    'latitude': latitude,
    'longitude': longitude,
    'start_date': start_date,
    'end_date': end_date,
    'hourly': ','.join(hourly_variables),
    'timezone': 'America/New_York'
}

response = requests.get(base_url, params=params)
```

```python
# Check if the request was successful

if response.status_code == 200:
    print('Data fetched successfully!')
else:
    print(f'Failed to fetch data. Status code: {response.status_code}')
```

```python
# Step 3: Load Data into pandas DataFrame

data = response.json()
hourly_data = data['hourly']

df = pd.DataFrame(hourly_data)
df.head()
```

```python
# Step 4: Data Cleaning

# Convert 'time' column to datetime
df['time'] = pd.to_datetime(df['time'])

# Set 'time' as the index
df.set_index('time', inplace=True)

# Check for missing values
df.isnull().sum()
```

```python
# Handle missing values (if any)
df.fillna(method='ffill', inplace=True)  # Forward fill
```

```python
# Step 5: Exploratory Data Analysis

# Summary Statistics
df.describe()
```

```python
# Plot Temperature Over Time

plt.figure(figsize=(14, 6))
plt.plot(df.index, df['temperature_2m'], label='Temperature (°C)')
plt.title('Temperature Over Time in New York City')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.legend()
plt.show()
```

```python
# Plot Relative Humidity Over Time

plt.figure(figsize=(14, 6))
plt.plot(df.index, df['relativehumidity_2m'], color='orange', label='Relative Humidity (%)')
plt.title('Relative Humidity Over Time in New York City')
plt.xlabel('Date')
plt.ylabel('Relative Humidity (%)')
plt.legend()
plt.show()
```

```python
# Plot Precipitation Over Time

plt.figure(figsize=(14, 6))
plt.bar(df.index, df['precipitation'], label='Precipitation (mm)')
plt.title('Precipitation Over Time in New York City')
plt.xlabel('Date')
plt.ylabel('Precipitation (mm)')
plt.legend()
plt.show()
```

```python
# Step 6: Correlation Analysis

# Calculate correlation between temperature and humidity
correlation = df['temperature_2m'].corr(df['relativehumidity_2m'])
print(f'Correlation between Temperature and Relative Humidity: {correlation:.2f}')
```

```python
# Scatter Plot of Temperature vs. Relative Humidity

plt.figure(figsize=(8, 6))
plt.scatter(df['temperature_2m'], df['relativehumidity_2m'], alpha=0.5)
plt.title('Temperature vs. Relative Humidity')
plt.xlabel('Temperature (°C)')
plt.ylabel('Relative Humidity (%)')
plt.show()
```

```python
# Step 7: Resample Data to Daily Averages

daily_avg = df.resample('D').mean()
daily_avg.head()
```

```python
# Plot Daily Average Temperature

plt.figure(figsize=(10, 5))
plt.plot(daily_avg.index, daily_avg['temperature_2m'], marker='o')
plt.title('Daily Average Temperature in New York City')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.grid(True)
plt.show()
```

```python
# Step 8: Compare with Another Location (Los Angeles)

# Coordinates for Los Angeles
latitude_la = 34.0522
longitude_la = -118.2437

params_la = {
    'latitude': latitude_la,
    'longitude': longitude_la,
    'start_date': start_date,
    'end_date': end_date,
    'hourly': ','.join(hourly_variables),
    'timezone': 'America/Los_Angeles'
}

response_la = requests.get(base_url, params=params_la)

# Load and clean Los Angeles data
data_la = response_la.json()
hourly_data_la = data_la['hourly']
df_la = pd.DataFrame(hourly_data_la)

df_la['time'] = pd.to_datetime(df_la['time'])
df_la.set_index('time', inplace=True)
df_la.fillna(method='ffill', inplace=True)
```

```python
# Resample Los Angeles Data to Daily Averages

daily_avg_la = df_la.resample('D').mean()
```

```python
# Combine DataFrames for Comparison

combined_temp = pd.DataFrame({
    'New York': daily_avg['temperature_2m'],
    'Los Angeles': daily_avg_la['temperature_2m']
})
```

```python
# Plot Comparison of Daily Average Temperature

combined_temp.plot(kind='bar', figsize=(10, 6))
plt.title('Daily Average Temperature: New York vs Los Angeles')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Step 9: Advanced Analysis - Temperature Difference

# Calculate Temperature Difference
combined_temp['Temp Difference'] = combined_temp['Los Angeles'] - combined_temp['New York']
combined_temp
```

```python
# Step 10: Save Data to CSV (Optional)

# Save New York data
df.to_csv('new_york_weather.csv')

# Save Los Angeles data
df_la.to_csv('los_angeles_weather.csv')
```

```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# for london
# Define parameters for London
latitude = 51.5074
longitude = -0.1278
start_date = "2022-01-01"
end_date = "2022-12-31"
timezone = "auto"

# Construct API URL
london_url = (
    f"https://archive-api.open-meteo.com/v1/archive?"
    f"latitude={latitude}&longitude={longitude}"
    f"&start_date={start_date}&end_date={end_date}"
    f"&hourly=precipitation&timezone={timezone}"
)

# Make the API request
london_response = requests.get(london_url)
london_data = london_response.json()
```

```python
# cleaning data for london
# Create DataFrame from London's hourly data
london_df = pd.DataFrame({
    'time': london_data['hourly']['time'],
    'precipitation': london_data['hourly']['precipitation']
})

# Convert time to datetime
london_df['time'] = pd.to_datetime(london_df['time'])

# Set time as index
london_df.set_index('time', inplace=True)

# Check for missing values
print(london_df.isnull().sum())
```

```python
# for seattle
#Define parameters for Seattle
latitude = 47.6062
longitude = -122.3321
start_date = "2022-01-01"
end_date = "2022-12-31"
timezone = "auto"

# Construct API URL
seattle_url = (
    f"https://archive-api.open-meteo.com/v1/archive?"
    f"latitude={latitude}&longitude={longitude}"
    f"&start_date={start_date}&end_date={end_date}"
    f"&hourly=precipitation&timezone={timezone}"
)

# Make the API request
seattle_response = requests.get(seattle_url)
seattle_data = seattle_response.json()
```

```python
# Data cleaning for seattlle
# Create DataFrame from Seattle's hourly data
seattle_df = pd.DataFrame({
    'time': seattle_data['hourly']['time'],
    'precipitation': seattle_data['hourly']['precipitation']
})

# Convert time to datetime
seattle_df['time'] = pd.to_datetime(seattle_df['time'])

# Set time as index
seattle_df.set_index('time', inplace=True)

# Check for missing values
print(seattle_df.isnull().sum())
```

```python
#resample
# Resample London and Seattle data by week (sum of precipitation)
london_weekly = london_df.resample('W').sum()
seattle_weekly = seattle_df.resample('W').sum()

# Preview the results
print("London weekly rainfall:")
print(london_weekly.head())

print("\nSeattle weekly rainfall:")
print(seattle_weekly.head())
```

```python
# visualize
# Plot the weekly rainfall totals
plt.figure(figsize=(14, 6))
plt.plot(london_weekly.index, london_weekly['precipitation'], label='London', color='blue')
plt.plot(seattle_weekly.index, seattle_weekly['precipitation'], label='Seattle', color='green')

# Chart formatting
plt.title("Weekly Rainfall in 2022: London vs Seattle")
plt.xlabel("Week")
plt.ylabel("Total Rainfall (mm)")
plt.legend()
plt.grid(True)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# OPtional
# Add a 4-week rolling average to smooth the lines
london_weekly['rolling_avg'] = london_weekly['precipitation'].rolling(window=4).mean()
seattle_weekly['rolling_avg'] = seattle_weekly['precipitation'].rolling(window=4).mean()

# Plot with rolling averages
plt.figure(figsize=(14, 6))
plt.plot(london_weekly.index, london_weekly['rolling_avg'], label='London (4-week Avg)', color='blue')
plt.plot(seattle_weekly.index, seattle_weekly['rolling_avg'], label='Seattle (4-week Avg)', color='green')

plt.title("Smoothed Weekly Rainfall Trends (2022)")
plt.xlabel("Week")
plt.ylabel("Rainfall (mm, 4-week average)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Analysis
The weekly precipitation data for 2022 reveals that Seattle consistently received significantly more rainfall than London, with Seattle’s average nearly double that of London. This disparity is clearly illustrated in the line plot, where Seattle shows both higher peaks and greater consistency in rainfall amounts.

Seattle’s rainfall peaks occurred during the first week of January and again in the final week of December, marking clear bookends to the year. In contrast, London displayed a smaller peak in early November, with relatively subdued rainfall throughout the rest of the year.

Between January and June, Seattle consistently outpaced London in total weekly precipitation. During the summer months (July through October), both cities experienced relatively dry conditions, though London exhibited slightly higher rainfall totals, particularly during August and September.

Rainfall activity increased for both cities from mid-October onward, extending into the winter months. This rise is captured well in the rolling average plot, which smooths out weekly variability and highlights seasonal trends more clearly.

Despite their reputations for gloomy, wet weather, the data confirms that Seattle experienced considerably more rainfall than London in 2022, both in terms of total accumulation and intensity of peak events.
